# Phase 2: Customer Journey & Incident Mapping (Python/Pandas)
**Objective:** To bridge siloed relational datasets and prove the causal link between driver incidents and resulting customer complaints.

```python
# Clone GitHub repository to get the dataset for this new notebook
!git clone https://github.com/gbzgbs/northstar_dataset.git
```

```python
import pandas as pd

# Load the datasets
df_complaints = pd.read_csv('/content/northstar_dataset/complaints.csv')
df_incidents = pd.read_csv('/content/northstar_dataset/incidents.csv')
df_deliveries = pd.read_csv('/content/northstar_dataset/deliveries.csv')

print("All datasets loaded successfully!")
```

```python
# 1. Merge Complaints with Deliveries (The Bridge) using 'order_id'
df_step_1 = pd.merge(df_complaints, df_deliveries, on='order_id', how='inner')

# 2. Merge the result with Incidents using 'delivery_id'
df_final_merged = pd.merge(df_step_1, df_incidents, on='delivery_id', how='inner')

# 3. Output suspicious cases where 'On Time' deliveries had hidden incidents
suspicious_cases = df_final_merged[['complaint_id', 'complaint_type', 'incident_type', 'delivery_status']]
suspicious_cases.head(10)
```

```python
import matplotlib.pyplot as plt

# Count the frequency of each incident type and plot it as a bar chart
suspicious_cases['incident_type'].value_counts().plot(kind='bar', color='coral')

# Add titles and labels for clarity
plt.title("Frequency of Hidden Incidents in 'On Time' Deliveries")
plt.xlabel("Incident Type")
plt.ylabel("Number of Occurrences")
plt.xticks(rotation=45) # Rotates the labels so they are easy to read
plt.show()
```
