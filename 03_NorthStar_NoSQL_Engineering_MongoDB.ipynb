```python
import pymongo
from pymongo import MongoClient

# Connect to the MongoDB Atlas Cluster using working credentials
mongo_uri = "mongodb+srv://uwltest:uwltest1@uwl.yj4rpyk.mongodb.net/?retryWrites=true&w=majority"
client = MongoClient(mongo_uri, tlsAllowInvalidCertificates=True)

# Define the database and collection
db = client['NorthStar_DB']
journey_collection = db['Customer_Journeys']

# 1. Create the Index (Query Optimisation)
print("Creating high-speed index on 'hub_id'...")
journey_collection.create_index([("delivery_details.hub_id", pymongo.ASCENDING)])
print("Index created successfully!\n")

# 2. Run the Analytical Compound Query
query = {
    "delivery_details.hub_id": "H08",
    "service_type": "Medical",
    "incidents": {"$exists": True, "$not": {"$size": 0}}
}

# 3. Execute and Output results
optimised_results = list(journey_collection.find(query))

print(f"REPORT: Found {len(optimised_results)} Medical deliveries at Hub H08 with active incidents.")

if len(optimised_results) > 0:
    print("\nSample Data for Report:")
    print(optimised_results[0])
```
